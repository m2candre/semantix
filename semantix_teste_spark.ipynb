```
# Create a RDD file
rdd = sc.textFile('/FileStore/tables/teste')
```

```
import re

def log_parser(line):

  # This method parse the log
  # The log format(standard) is Apache Common Log Format (CLF): https://httpd.apache.org/docs/1.3/logs.html#common
  # Example:
  # in24.inetnebr.com - - [01/Aug/1995:00:00:01 -0400] "GET /shuttle/missions/sts-68/news/sts-68-mcc-05.txt HTTP/1.0" 200 1839

  # Compile regex pattern for better performance
  regex_pattern = re.compile('^(\S+) (\S) (\S) \[(\d{2}/\w{3}/\d{4}:\d{2}:\d{2}:\d{2}) (\+|-\d{4})\] \"(\S+)\s*(\S+) (\w+)/(\d{1}.\d{1})\" (\d{3}) (\d+)')
  
  match = regex_pattern.match(line)

  if match:
      parsed = [match.groups()[0],
                match.groups()[3],
                match.groups()[4],
                match.groups()[5],
                match.groups()[6],
                match.groups()[7],
                match.groups()[8],
                int(match.groups()[9]),
                int(match.groups()[10])]
      return parsed
  else:
    print(line)
```

```
# Convert RDD to DataFrame
from pyspark.sql.types import StructField, StructType, StringType, ShortType, LongType

fields = [StructField('host', StringType(), True),
         StructField('datetime', StringType(), True),
         StructField('timezone', StringType(), True),
         StructField('request_method', StringType(), True),
         StructField('url', StringType(), True),
         StructField('protocol', StringType(), True),
         StructField('protocol_version', StringType(), True),
         StructField('status_code', ShortType(), True),
         StructField('bytes_transferred', LongType(), True)]

schema = StructType(fields)

parsed_df = sqlContext.createDataFrame(parsed_rdd, schema)
```

```
# 1. Número de hosts únicos.
parsed_df.select('host').distinct().count()
```

```
# 2. O total de erros 404.
parsed_df.filter(parsed_df['status_code'] == '404').count()
```

```
# 3. Os 5 URLs que mais causaram erro 404.
df_error_404 = parsed_df.filter(parsed_df['status_code'] == '404').groupBy('host').count().orderBy('count', ascending=0).take(5)

# Como a amostra é pequena e não possui o 404, ao realizar o display em uma lista vazia irá causar erro, então foi necessário realizar esse teste antes
if df_error_404:
  display(df_error_404)
```
